
# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
import pandas as pd
a = pd.DataFrame({
    "City" : ['Bhopal','Indore','Bhopal','Delhi']
})
print(a["City"].unique().tolist())
````


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
import pandas as pd
b=pd.DataFrame({
    "City" : ['Bhopal','Bhopal','Delhi'],
    "Marks": [80,90,70]
})
print(b.groupby('City')['Marks'].mean())
````


# Q5. Sort Values

## Problem Statement
Sort students by Marks in ascending order.

## Sample Output
```python
Sorted DataFrame
```


````python
student = pd.DataFrame({
    "Name": ['anuj','Ankit','Aman','Deepak'],
    "Marks":[87,56,98,43]

})
print(student.sort_values('Marks',ascending= True))
````


# Q6. Sort Values Descending

## Problem Statement
Sort students by Marks in descending order.

## Sample Output
```python
Sorted DataFrame
```


````python
student = pd.DataFrame({
    "Name": ['anuj','Ankit','Aman','Deepak'],
    "Marks":[87,56,98,43]

})
print(student.sort_values('Marks',ascending = False))
````


# Q7. Filter Records

## Problem Statement
Display students having marks greater than 75.

## Sample Output
```python
Filtered rows
```


````python
student = pd.DataFrame({
    "Name": ['anuj','Ankit','Aman','Deepak'],
    "Marks":[87,56,98,43]

})
print(student[student['Marks'] > 75].sort_values('Marks'))
````


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
student = pd.DataFrame({
    "Name": ['anuj','Ankit','Aman','Deepak'],
    "Marks":[87,56,98,43],
    "Age":[16,21,20,18]

})
print(student[student['Marks'] >70 & (student['Age'] >20)].sort_values(['Marks','Age']))
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
data = pd.DataFrame({
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000]
})
print(data.groupby('Department')['Salary'].sum())
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
student = pd.DataFrame({
    "Name": ['anuj','Ankit','Aman','Deepak'],
    "Marks":[87,56,98,43],
    "Age":[16,21,20,18],
    "City":['Mumbai','Pune','Delhi','Pune']
})

print(student["City"].unique().tolist())
print(student['City'].value_counts())
print(student[student['Marks']>80].sort_values('Marks'))
print(student.sort_values('Marks',ascending =True))
````
