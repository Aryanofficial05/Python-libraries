## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
import pandas as pd
a = pd.DataFrame({
    "Age":[25,18,30,22]
})
print(a.sort_values("Age",ascending = True))
```

## Q2. Sort Marks in Descending Order

Input:
Marks = [70, 95, 80, 60]

Output:
95
80
70
60

```python
b = pd.DataFrame({
    "Marks":[70,95,80,60]
})
print(b.sort_values("Marks",ascending=False))
```

## Q3. Find Mean Age

Input:
Age = [20, 25, 30, 25]

Output:
25.0

```python
a = pd.DataFrame({
    "Age":[20,25,30,25]
})
print(a['Age'].mean())
```

## Q4. Find Median Marks

Input:
Marks = [50, 60, 70, 80, 90]

Output:
70

```python
b = pd.DataFrame({
    "Marks":[50,60,70,80,90]
})
print(b['Marks'].median())
```

## Q5. Find Mode

Input:
Age = [20, 25, 25, 30, 25]

Output:
25

```python
a = pd.DataFrame({
    "Age":[20,25,25,30,25]
})
print(a['Age'].mode())
```

## Q6. Find Maximum Score

## Q7. Find Minimum Score

Input:
Score = [78, 90, 67, 88]

Output:
67

```python
d = pd.DataFrame({
    "Score":[78,90,67,88]
})
print(d['Score'].min())
```

## Q8. Count City Frequency

Input:
City = ['Bhopal','Indore','Bhopal','Jabalpur']

Output:
Bhopal      2
Indore      1
Jabalpur    1

```python
a = pd.DataFrame({
    "City":['Bhopal','Indore','Bhopal','Jabalpur']
})
print(a['City'].value_counts())
```

## Q9. Find Percentage Frequency

Input:
City = ['A','A','B','C']

Output:
A = 50%
B = 25%
C = 25%

```python
a = pd.DataFrame({
    "City":['A','A','B','C']
})
print(a["City"].value_counts(normalize= True) * 100)
```

## Q10. Sort DataFrame by Salary

Input:
Name   Salary
Aman   30000
Raj    50000
Riya   40000

Output:
Raj
Riya
Aman

```python
a = pd.DataFrame({

    "Name":['Aman','Raj','Riya'],
    "Salary":[30000,50000,60000]

})
print(a['Name'].sort_values())
```

## Q11. Find Correlation

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
1.0

```python
a = pd.DataFrame({
    "Math":[10,20,30,40],
    "Science":[15,25,35,45]
})
print(a['Math'].corr(a['Science']))
```

## Q12. Find Covariance

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
166.67 (approx)

```python
a = pd.DataFrame({
    "Math":[10,20,30,40],
    "Science":[15,25,35,45]
})
print(a['Math'].cov(a['Science']))
```

## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
b = pd.DataFrame({
    "Marks":[60,70,80,90]
})
print(b['Marks'].agg(['mean','median','min','max']))
```
